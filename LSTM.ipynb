```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os

# Define the input image dimensions and other hyperparameters
img_size = (64, 64)
batch_size = 32

# Create a data generator
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)  # Split 20% of data for validation

# Define the data generator for training and validation
dataset_directory = 'achulu'  # Replace with your dataset directory
train_generator = datagen.flow_from_directory(
    dataset_directory,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True,
    seed=42,
    subset='training'  # This is the training subset
)

validation_generator = datagen.flow_from_directory(
    dataset_directory,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False,  # Set shuffle to False for the validation subset
    seed=42,
    subset='validation'  # This is the validation subset
)

# Mapping of directory names to class labels
class_mapping = {
    'a': 0,
    'aa': 1,
    'ah': 2,
    'ai': 3,
    'am': 4,
    'ao': 5,
    'E': 6,
    'Ee': 7,
    'I': 8,
    'Ii': 9,
    'O': 10,
    'Oo': 11,
    'Ru': 12,
    'Ruu': 13,
    'U': 14,
    'Uu': 15,
    # Add more classes as needed
}

# Define the model
model = keras.Sequential([
    layers.Input(shape=(img_size[0], img_size[1], 3)),  # Adjusted input shape
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(len(class_mapping), activation='softmax'),
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(train_generator, epochs=10, validation_data=validation_generator)

# Save the model
model.save('sequence_model.h5')
```

```python
# Define the data generator for validation
validation_generator = datagen.flow_from_directory(
    'achulu',
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False,  # Set shuffle to False for validation
    seed=42,
    subset='validation'
)

# Evaluate the model on the validation dataset
validation_results = model.evaluate(validation_generator)

# Print the accuracy
accuracy = validation_results[1]  # The second element of the evaluation results is accuracy
print("Validation Accuracy: {:.2f}%".format(accuracy * 100))

```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing import image
import numpy as np

# Load the trained model
model = keras.models.load_model('sequence_model.h5')

# Load an image you want to classify
img_path = 'LSTM_testing.jpg'  # Replace with the path to your image
img = image.load_img(img_path, target_size=(64, 64))  # Adjust the target size

# Preprocess the image
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array /= 255.0  # Normalize the image

# Make predictions
predictions = model.predict(img_array)

# Get the predicted class index
predicted_class_index = np.argmax(predictions, axis=1)[0]

# Get the class label from the mapping
class_mapping = {
    0: 'a',
    1: 'aa',
    2: 'ah',
    3: 'ai',
    4: 'am',
    5: 'ao',
    6: 'E',
    7: 'Ee',
    8: 'I',
    9: 'Ii',
    10: 'O',
    11: 'Oo',
    12: 'Ru',
    13: 'Ruu',
    14: 'U',
    15: 'Uu',
    # Add more classes as needed
}

predicted_class_label = class_mapping[predicted_class_index]

# Print the predicted class label
print(f'Predicted Class: {predicted_class_label}')

```
